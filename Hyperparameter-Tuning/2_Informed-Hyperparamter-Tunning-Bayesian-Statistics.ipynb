#### TPOT - Genetic Hyperparameter Tuning

- Python Automated Machine Learning tool that optimizes machine learning pipelines using genetic programming.

Pipeline not only includes the model (or multiple models) BUT also works on features and other aspects of the process.  Plust it returns the Python code of the pipeline for you!

key arguments:
- generations: iterations to run training for
- population_size: number of models to keep after each iteration
- offspring_size: number of models to produce in each iteration
- mutation_rate: proportion of pipelines to apply randomness to
- crossover_rate: proportion of pipelines to breed each iteration
- scoring: the function to determine the best models
- cv: cross-validation strategy to use
